```python
import numpy as np
array=np.array([9,8,7,6,5,4,3,2,1,0])
new_array=array.reshape(5,2)
print(new_array)
```

```python
import numpy as np
arr=np.array([[1,2,9,3,4],[5,6,7,0,8]])
new_array=arr.transpose()
print(new_array)
```
