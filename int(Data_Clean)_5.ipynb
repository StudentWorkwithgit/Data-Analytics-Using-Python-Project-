```python
import pandas as pd
data={
    "Name":["beevi","vigi","kavi","jana","vino"],
    "Sales":[12000,300,18000,15000,None]
}
df=pd.DataFrame(data)
print(df)
```

```python
print(df.isnull())
```

```python
print(df.dropna())
```

```python
clear_df=df.dropna()
print(clear_df)
```

```python
print(df.fillna(0))
```

```python
df.fillna(0,inplace=True)
print(df)
```

```python
import pandas as pd
data={
    "Marks":[98,90,87,None,96]
}
df=pd.DataFrame(data)
mean_value=df["Marks"].mean()
df["Marks"].fillna(mean_value,inplace=True)
print(df)
```
