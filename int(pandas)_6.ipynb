```python
import pandas as pd
Employees={
    "Em_ID":[1001,1002,1003,1004,1005],
    "Name":["Bavi","Kavi","Nivi","Est","William"],
    "Department":["IT","HR","IT","FINANCE","MARKETING"],
    "Salary":[40000,41000,500000,440000,440000],
    "Experience": [1, 3, 5, 2, 2]

}
df = pd.DataFrame(Employees)

print("Employees Dataset")
print(df)

print("\nAverage Salary")
print(df["Salary"].mean())

print("\nHighest Salary")
print(df["Salary"].max())

print("\nEmployees in IT")
print(df[df["Department"] == "IT"])
```

```python
import pandas as pd
data={
    "Emp":["Moomina","Kaviya","Majitha","Moomina"],
    "Salary":[1000000,1000000,1200000,1000000]
}
df=pd.DataFrame(data)
df=df.drop_duplicates()
print(df)
```

```python
import pandas as pd
data={
    "Emp":["Moomina","Kaviya","Majitha"],
    "Sal":[1000000,1500000,1200000]
}
df=pd.DataFrame(data)

df.rename(columns={
    "Emp":"Employee Name",
    "Sal":"Employee Salary"
},inplace=True)
print(df)
```

```python
import pandas as pd
data={
    "Emp":["Moomina","Kaviya","Majitha","Asila",None,"Sahana","Safri",None,"Moomina"],
    "Sal":[1000000,1200000,490000,1500000,1200000,1100000,780000,None,1000000]
}
df=pd.DataFrame(data)
df=df.fillna(0)
print(df)

df.rename(columns={
    "Emp":"Employee Name",
    "Sal":"Employee Salary"
},inplace=True)
print(df)

print(df.duplicated())
df=df.drop_duplicates()
print(df)
```
