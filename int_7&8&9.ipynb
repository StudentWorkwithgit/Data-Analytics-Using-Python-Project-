```python
import pandas as pd
data={
    "Position":["HR","IT","HR","IT","MARKETING"],
    "Age":[20,21,18,21,22],
    "Sal":[1000000,1000000,130000,90000,1200000]
}
df=pd.DataFrame(data)
print(df)
```

```python
Same_Pos=df[df["Position"]=="HR"]
print(Same_Pos)
```

```python
high_sal=df[df["Sal"]>100000]
print(high_sal)
```

```python
sorted_df=df.sort_values("Sal")
print(sorted_df)
```

```python
Position_Grp=df.groupby("Position")["Sal"].sum()
print(Position_Grp)
```

```python
Age_Grp=df.groupby("Age")["Sal"].sum()
print(Age_Grp)
```

```python
from google.colab import files
import pandas as pd

uploaded=files.upload()
df = pd.read_csv('Sales.csv')

df_active = df[df['Sales'] > 0]

df_sorted = df_active.sort_values(
    ['Country', 'Sales'],
    ascending=[True, False]
)

report = df_sorted.groupby('Country')['Sales'].agg(
    ['sum', 'mean', 'count']
)
report.columns = [
    'Total Sales',
    'Avg Sale',
    'Order Count'
]

print(report)
```

```python
import matplotlib.pyplot as plt

years=[2021,2022,2023,2024,2025,2026]
sales=[12000,13000,5000,33000,37000,56000]

plt.plot(years,sales)
plt.title("Yearly Sales")
plt.xlabel("years")
plt.ylabel("Sales Amount")
plt.show()
```

```python
from matplotlib.lines import lineStyles
import matplotlib.pyplot as plt

years=[2021,2022,2023,2024,2025,2026]
sales=[12000,13000,5000,33000,37000,56000]

plt.plot(years,sales,color="maroon",marker="o",linestyle="--")
plt.title("Yearly Sales")
plt.xlabel("years")
plt.ylabel("Sales Amount")
plt.show()
```

```python
import matplotlib.pyplot as plt

years=[2021,2022,2023,2024,2025,2026]
sales=[12000,13000,5000,33000,37000,56000]

plt.bar(years,sales,color="red")
plt.title("Yearly Sales")
plt.xlabel("years")
plt.ylabel("Sales Amount")
plt.show()
```
