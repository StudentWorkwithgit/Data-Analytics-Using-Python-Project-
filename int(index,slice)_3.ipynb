```python
Marks=[98,90,97,87,78]

Total=sum(Marks)
Avg=Total/5
Max=max(Marks)
Min=min(Marks)

print("Total:",Total)
print("Average:",Avg)
print("Max Value:",Max)
print("Min Value:",Min)
```

```python
Sales=[10000,18000,90000,68000,99000]

total=sum(Sales)
Avg=total/len(Sales)
high_Sales=max(Sales)
low_Sales=min(Sales)

print("Total:",total)
print("Average:",Avg)
print("high_Sales:",high_Sales)
print("low_Sales:",low_Sales)
```
